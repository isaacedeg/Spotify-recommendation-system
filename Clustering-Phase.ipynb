```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import plotly.express as px
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('Data/spotify-data.csv')
genre_column = data['genre']
year_column = data['year']
artist_column = data['artist_name']
data.dropna(inplace=True)
year_data = pd.DataFrame(year_column)
genre_data = pd.DataFrame(genre_column)
artist_data = pd.DataFrame(artist_column)
# Function to get decade from year
def get_decade(year):
    return str(year // 10 * 10) + 's'

# Apply function to create decade column
data['Decade'] = data['year'].apply(lambda year: get_decade(year))
```

```python
# Extract features for clustering
df = pd.read_csv('Data/first_10000_rows.csv')
features = df.drop(columns=['genre', 'track_id', 'Decade', 'artist_name', 'track_name', 'popularity', 'year', 'mode', 'duration_ms', 'time_signature'], axis=1)
# Fit K-means clustering model
kmeans = KMeans(n_clusters=12, random_state=42)

kmeans.fit(features)
df['cluster'] = kmeans.predict(features)

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(features)

# Add t-SNE results to the DataFrame
df['tsne-1'] = tsne_results[:, 0]
df['tsne-2'] = tsne_results[:, 1]

# Plotting with plotly for interactive visualization
fig = px.scatter(
    df, x='tsne-1', y='tsne-2',
    color='cluster', hover_data=['genre'],
    title="TSNE Clustering of Genres"
    )

fig.update_layout(
    xaxis_title="TSNE Dimension 1",
    yaxis_title="TSNE Dimension 2"
)

fig.show()
```

```python
df = pd.read_csv('Data/first_10000_rows.csv')
# Extract features for clustering
features = df.drop(columns=['genre', 'track_id', 'Decade', 'artist_name', 'track_name', 'popularity', 'year', 'mode', 'duration_ms', 'time_signature'], axis=1)

# Fit K-means clustering model
kmeans = KMeans(n_clusters=25, random_state=42)
kmeans.fit(features)
df['clusters'] = kmeans.predict(features)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
pca_results = pca.fit_transform(features)

# Add PCA results to the DataFrame
df['pca-one'] = pca_results[:, 0]
df['pca-two'] = pca_results[:, 1]

# Plotting with plotly for interactive visualization
fig = px.scatter(
    df, x='pca-one', y='pca-two',
    color='clusters', hover_data=['track_name'],
    title="PCA Clustering of Songs"
)

fig.update_layout(
    xaxis_title='PCA Dimension 1',
    yaxis_title='PCA Dimension 2'
)

fig.show()
```
